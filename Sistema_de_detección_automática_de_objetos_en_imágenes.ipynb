# Introducción

La visión por computadora es una rama de la inteligencia artificial que permite a las máquinas interpretar y analizar imágenes y videos. Uno de los problemas más importantes en este campo es la detección de objetos, cuyo objetivo es identificar y localizar múltiples objetos dentro de una imagen.

En este proyecto implemento un sistema de detección de objetos utilizando el modelo YOLOv8, el cual permite realizar detección en tiempo real con alta precisión. La implementación se realiza en Google Colab, aprovechando el uso de GPU para acelerar el procesamiento.

El sistema es capaz de detectar objetos como personas, vehículos, animales y otros elementos comunes presentes en el conjunto de datos COCO.


# Objetivo

Desarrollar un sistema de inteligencia artificial que detecte automáticamente objetos en imágenes utilizando un modelo de aprendizaje profundo preentrenado, permitiendo visualizar los resultados mediante cajas delimitadoras y etiquetas de clasificación.

La librería Ultralytics contiene implementaciones optimizadas del modelo YOLOv8, facilitando su uso para tareas de detección, segmentación y clasificación de imágenes.


```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```

Se utiliza el modelo YOLOv8n, el cual es una versión liviana del modelo YOLOv8 preentrenada con el conjunto de datos COCO, que contiene más de 80 clases de objetos diferentes.


```python
from google.colab import files
uploaded = files.upload()
```

El usuario puede cargar una imagen desde su computadora para que el sistema realice la detección de objetos automáticamente.


```python
image_name = list(uploaded.keys())[0]
results = model.predict(source=image_name, save=True)
```

El modelo analiza la imagen y genera una predicción que incluye:
- Coordenadas de los objetos detectados
- Probabilidad de cada clase
- Etiqueta correspondiente al objeto identificado


```python
from IPython.display import Image
Image(filename="runs/detect/predict/" + image_name)
```

La imagen resultante muestra las cajas delimitadoras (bounding boxes) alrededor de cada objeto detectado junto con su etiqueta y nivel de confianza.


# Interpretación de resultados

Cada objeto detectado es representado mediante una caja rectangular y una etiqueta que indica la clase del objeto. El porcentaje representa la confianza del modelo en la predicción realizada.

Este tipo de sistema es utilizado en aplicaciones como:
